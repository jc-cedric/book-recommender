```python
import kagglehub
from seaborn import heatmap
from traitlets.utils import descriptions

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax  = plt.axes()
sns.heatmap(books.isna().transpose(),cbar=False,ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing values")

plt.show()
```

**Interpretation**

We can see that:
- the *subtitle* variable has the most significant missing values,
- the variables *categories*, *thumbnail*, and *description* also have missing values but their missing values seems to be randomly scattered
- the variables *average_rating*, *num_pages* and *ratings_count*, missing values present a similarity pattern

```python
import numpy as np
```

```python
books["missing_description"] = np.where(books["description"].isna(), 1,0)
books["age_of_book"] = 2025 - books["published_year"]
```

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

# pearson correlation are more appropriate when dealing with completely continuous values
# spearman correlation are more appropriate when dealing with non-continuous values
correlation_matrix = books[columns_of_interest].corr(method="spearman")

sns.set_theme(style="white")
plt.figure(figsize=(8,6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={"label": "Spearman correlation"})
heatmap.set_title("Correlation heatmap")
plt.show()
```

```python
# Estimating the number of rows with missing values
books[(books["description"].isna()) |
      (books["num_pages"].isna()) |
      (books["average_rating"].isna()) |
      (books["published_year"].isna())
]
```

**Interpretation:** We have about 300 rows with missing values which is less than 10% of our dataset, that means we can get rid of them

```python
books_missing = books[~(books["description"].isna()) &
      ~(books["num_pages"].isna()) &
      ~(books["average_rating"].isna()) &
      ~(books["published_year"].isna())
]
```

```python
books_missing
```

**Observation:** We have too many categories

```python
# Doing this to have an overview of the categories distribution in our dataset
books_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

**Conclusion:** We have an uneven distribution of the categories in our dataset

**Remark** By inspecting the *descrpition* column we can notice that some books have non meaningful descriptions.

```python
books_missing_25_words = books_missing[books_missing["words_in_description"] >= 25]
```

```python
books_missing_25_words
```

```python
books_missing_25_words["title_and_subtitle"] = (
    np.where(books_missing_25_words["subtitle"].isna(), books_missing_25_words["title"],books_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)
```

```python
books_missing_25_words
```

```python
books_missing_25_words["tagged_description"] = books_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
```

```python
#end of data preprocessing
#dropping the non relevant columns
(
    books_missing_25_words.drop(columns = ["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1)
    .to_csv("./data/books_cleaned.csv", index=False)
)
```
